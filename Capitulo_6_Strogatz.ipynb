# **CAPITULO 6 STROGATZ - NONLINEAR DYNAMICS AND CHAOS**

En este cuaderno se observaran algunos de los ejercicios ejemplos y propuesto correspondientes al capitulo 6 del libro de Strogatz


$$\dot x=-x+x^3$$
$$\dot y= -2y $$



```
import numpy as np 
import matplotlib.pyplot as plt 
from sympy import var, solve, Matrix,det,sin,nonlinsolve,Lambda,ImageSet,pi

eje = np.arange(-2,2,0.01)
x = eje
y = eje
X,Y = np.meshgrid(x,y)

# encontrar las ra√≠ces 
x, y = var('x y')

f1 = -x + x**3
f2 = -2*y 

sols = nonlinsolve((f1, f2), (x, y))

f = Matrix((f1,f2))
jacobiano = f.jacobian((x,y)).T

# clasificando los puntos con la ayuda de los valores propios 
g = []
for sol in sols:
    matrix =(jacobiano.subs([(x,sol[0]),(y,sol[1])]))
    l1 = var('l_1')
    d=det((matrix-l1*Matrix([[1,0], [0,1]])))
    h = solve(d,l1)
    if h[0]<0 and h[1] <0:
        g.append('r')
    else:
        g.append('b')

#encontrar los flujos 
x_dot = -X + X**3
y_dot = -2*Y

plt.figure(figsize=(10,5))

plt.streamplot(X,Y,x_dot,y_dot,density = 1,linewidth=1,minlength=0.2)
for i,sol in enumerate(sols):
    plt.plot(sol[0],sol[1],'o'+g[i],label="equilibrio")

plt.grid()
plt.xlabel('$\dot{x}$')
plt.ylabel('$\dot{y}$')
plt.legend(loc='best')
plt.show()
```

$$ \dot x=x(3-x-2y)\dot y=y(2-x-y) $$

```

X,Y = np.meshgrid(np.arange(0,3.5,0.01),np.arange(0,2.5,0.01))

# encontrar las ra√≠ces 
x, y = var('x y')

f1 = x*(3-x-2*y)
f2 = y*(2-x-y) 

sols = nonlinsolve((f1, f2), (x, y))

f = Matrix((f1,f2))
jacobiano = f.jacobian((x,y)).T

# clasificando los puntos con la ayuda de los valores propios 
g = []
for sol in sols:
    matrix =(jacobiano.subs([(x,sol[0]),(y,sol[1])]))
    l1 = var('l_1')
    d=det((matrix-l1*Matrix([[1,0], [0,1]])))
    h = solve(d,l1)
    if h[0]<0 and h[1] <0:
        g.append('r')
    else:
        g.append('b')

#encontrar los flujos 
x_dot = X*(3 - X - 2*Y)
y_dot = Y*(2- X - Y)

plt.figure(figsize=(10,5))

plt.streamplot(X,Y,x_dot,y_dot,density =3,linewidth=1,minlength=0.2)
for i,sol in enumerate(sols):
    plt.plot(sol[0],sol[1],'o'+g[i],label="equilibrio")

plt.grid()
plt.xlabel('$\dot{x}$')
plt.ylabel('$\dot{y}$')
plt.legend(loc='best')
plt.show()
```
