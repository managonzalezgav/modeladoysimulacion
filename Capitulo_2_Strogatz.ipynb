#Ecuación logística

$$ \dot N=rN(1- \frac{N}{K})  $$


```
import numpy as np
import matplotlib.pyplot as plt

N = np.arange(0,12,.1)
r,k = 1,10
dN = r*N*(1-N/k)


fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_xticks([5,10])
ax.set_yticks([0,2])

plt.plot(N,dN)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def ECL(t, N, r, k): 
  return  r*N*(1-N/k)

N0 = np.arange(2,20,2)
r,k = 1,10
tspan = [0, 15]
sol = solve_ivp(ECL, tspan, N0, args = (r,k),\
                dense_output = True)

t = np.linspace(tspan[0],tspan[1],200)
x = sol.sol(t)
for i in range(N0.size):
  plt.plot(t,x[i])
plt.show()
```

#EJERCICIO 2.2.1

$$ \dot x=4x^2-16 $$


```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-6,6,.1)
dx = 4*(x**2)-16


fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_xticks([-4,-2,2,4])
ax.set_yticks([-20,20,40,60])

plt.plot(x,dx)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def dx(t, x, c, d): 
   return c*(x**2)-d

a,b = 4,16
x0 = np.array([-6, -4, -3, -2,-1,0, 1, 1.5, 1.8, 2, 2.0001])
tspan = [0, 0.6]
sol = solve_ivp(dx, tspan, x0,args=(a,b),\
                dense_output = True)

t = np.linspace(tspan[0],tspan[1],101)
x = sol.sol(t)
for i in range(x0.size):
  plt.plot(t,x[i])
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def dx(t, x): 
  return  1-x**14

x0 = np.arange(-0.900000,1.40000,0.1)
tspan = [0, 10.809]
sol = solve_ivp(dx, tspan, x0, \
                dense_output = True)

t = np.linspace(tspan[0],tspan[1],101)
x = sol.sol(t)
for i in range(x0.size):
  plt.plot(t,x[i])
plt.show()
```

#EJERCICIO 2.2.4

$$ \dot x=e^{-x}*sin(x) $$

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-np.pi,np.pi*2,.01)
dx = np.exp(-x)*np.sin(x)


fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

plt.plot(x,dx)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def dx(t, x): 
  return  np.exp(-x)*np.sin(x)

x0 = np.arange(-3*np.pi,3*np.pi,np.pi/4)
tspan = [0, 10000]
sol = solve_ivp(dx, tspan, x0, \
                method = 'Radau', dense_output = True)

t = np.linspace(tspan[0],tspan[1],101)
x = sol.sol(t)
for i in range(x0.size):
  plt.plot(t,x[i])
plt.show()
```

#EJERCICIO 2.2.7

$$\dot x=e^x-cos(x)  $$

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10*np.pi,np.pi/4,.1)
dx = np.exp(x)-np.cos(x)


fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.plot(x,dx)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def dx(t, x): 
  return  np.exp(x)-np.cos(x)

x0 = np.arange(-3*np.pi+0.0001,0*np.pi/4+.0002,np.pi/4)
tspan = [0, 8.8]
sol = solve_ivp(dx, tspan, x0, \
                method = 'Radau', dense_output = True)
t = np.linspace(tspan[0],tspan[1],101)
x = sol.sol(t)
for i in range(x0.size):
  plt.plot(t,x[i])
plt.show()
```
