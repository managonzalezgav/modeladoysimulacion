#2.2. Bifurcaciones de adición de período
En este fenómeno , las ventanas de períodos consecutivos
están separados por regiones de caos. En otras palabras, como el parámetro es variado, se obtiene una órbita estable de período-n  , n = 1, 2,... . ...seguido de una región de caos, luego una órbita de período estable (n + 1), seguida de un caos, y luego una órbita de período (n+2) y
etcétera. En las siguientes figuras 2.1 a las 2.5, se muestran ejemplos de una de estas secuencias.



```
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp



def fx(x,a,b):
  return b*x + (1/2)*(a-b)*(np.abs(x+1)-np.abs(x-1))

def chua (t,y,alpha,beta,gamma,k,a,b):
   return [k*alpha*(y[1]-y[0]-fx(y[0],a,b)), k*(y[0]-y[1]+y[2]) , k*(-beta*y[1]-gamma*y[2]) ]

alpha=3.70782609
beta=3.79752551
gamma=0.08015844
a=-1.285098
b=-0.6029288
k=1

tin,tfin=0,50

t= np.linspace(tin,tfin,10000)
y0=[0,-0.125,-0.646]

sol=solve_ivp(chua,[tin,tfin],y0,args=(alpha,beta,gamma,k,a,b),t_eval=t)

ax = plt.axes(projection='3d')
ax.plot3D(sol.y[0], sol.y[1], sol.y[2])
ax.scatter3D(sol.y[0][-1], sol.y[1][-1], sol.y[2][-1],'o',color='red')
#ax.scatter3D(,"o",color="black")

plt.title('FÍGURA 2.1')
plt.show()
```

#2.3. Órbitas homoclínicas y heteroclínicas en el oscilador de Chua
Estrechamente relacionado con la aparición de un comportamiento caótico en los sistemas dinámicos en
en general son las llamadas trayectorias homoclínicas y heteroclínicas. Una trayectoria homoclínica
es aquella cuyo punto límite, tanto en el tiempo hacia adelante como hacia atrás, es el mismo punto de equilibrio de tipo silla de montar. Por otro lado, dos puntos de equilibrio diferentes
son los puntos límite en el tiempo hacia adelante y hacia atrás, respectivamente, para una trayectoria, ver las dos siguientes fíguras:

#2.4. Fenómenos de coexistencia
En general, para el mismo conjunto de parámetros, existen muchos límites estables e inestables establecidos. La trayectoria del sistema convergerá hacia un atractor particular si las
las condiciones iniciales se eligen en la cuenca de atracción del atractor. Así, ¿cuál de las
los atractores coexistentes que observamos en los experimentos depende del estado inicial de la
sistema?. Pues se puede decir que todos los atractores anteriormente mostrados dependen de las condiciones iniciales que se establezcan en el sistema. La coexistencia de atrayentes es un fenómeno interesante en el que la interacción
de los atractores pueden dar lugar a diferentes fenómenos dinámicos
